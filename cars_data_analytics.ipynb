```python
# ========================
# ðŸš— Cars Dataset Analysis
# ========================

# ðŸ“Œ Step 1: Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Configure plots
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)

# ========================
# ðŸ“Œ Step 2: Load Dataset
# ========================
df = pd.read_csv("/content/cars.csv")

print("Shape of dataset:", df.shape)
df.head()
```

```python
# ðŸ“Œ Step 3: Explore Dataset
print("\nDataset Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# ========================
# ðŸ“Œ Step 4: Visualizations
# ========================

# 1. Distribution of Car Prices (Histogram + KDE)
plt.figure()
sns.histplot(df['Price'], kde=True, bins=30, color="blue")
plt.title("Distribution of Car Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()

# 2. Car Prices by Fuel Type (Boxplot)
plt.figure()
sns.boxplot(x="Fuel_type", y="Price", data=df, palette="Set2")
plt.title("Car Prices by Fuel Type")
plt.show()

# 3. Engine Size vs Price (Scatterplot)
plt.figure()
sns.scatterplot(x="Engin_size", y="Price", hue="Fuel_type", data=df, alpha=0.7)
plt.title("Engine Size vs Price")
plt.show()

# 4. Horsepower vs Price (Regression Plot)
plt.figure()
sns.regplot(x="Engine_power", y="Price", data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title("Horsepower vs Price")
plt.show()

# 5. Price by Transmission (Violin Plot)
plt.figure()
sns.violinplot(x="Gearbox", y="Price", data=df, palette="muted")
plt.title("Car Prices by Transmission Type")
plt.show()

# 6. Average Price by Fuel Type (Bar Plot)
plt.figure()
sns.barplot(x="Fuel_type", y="Price", data=df, estimator="mean", ci=None, palette="coolwarm")
plt.title("Average Car Price by Fuel Type")
plt.show()
```
